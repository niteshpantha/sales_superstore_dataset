```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# reading csv files and displaying first five entries
df = pd.read_csv("/home/nitesh/Downloads/sales_superstore_dataset/train.csv")
df[:5]
```

```python
# displaying the features
df.columns.to_list()
```

```python
# displaying necessary information of the features
df.info()
```

```python
# checking the missing values and sorting ascending order and displaying top 5 missing values
df.isna().sum().sort_values(ascending=False)[:5]
```

```python
# imputing feature with mode
mode_value = df['Postal Code'].mode()[0]
df['Postal Code'].fillna(mode_value, inplace=True)
```

```python
# checking is missing values still persists
df.isna().sum().sort_values(ascending=False)[:5]
```

```python
df['Order Date'][:5]
```

```python
df['Ship Date'][:3]
```

```python
# changing features to appropriate formats
df_dates = ['Order Date', 'Ship Date']
for col in df_dates:
    df[col] = pd.to_datetime(df[col], format='%d/%m/%y', errors='coerce')
```

```python
df.info()
```

```python
# storing the cleaned file into csv format on same directory
df.to_csv("/home/nitesh/Downloads/sales_superstore_dataset/cleaned_train.csv", index=False)
```
